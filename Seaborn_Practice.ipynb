## Seaborn: 

 It is a python library which is used for data visualisation and making plots. 

```python
#importing the library

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Seaborn has some in builts datasets to practice the plots. 

```python
# total bill vs tip dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for our plots

sns.set_theme()
```

```python
#visualise the data

sns.relplot(data= tips, x = 'total_bill', y='tip', col='time', hue='smoker', style='smoker', size= 'size')
```

```python
# house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()

house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
house.head()
```

HEAT MAP: It is drawn to visualise the correlation between the various features. 

```python
correlation = house.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(data = correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
